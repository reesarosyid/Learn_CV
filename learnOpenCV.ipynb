```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
import cv2
```

## Read Image

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

Why result image above has not correct color? 

It because opencv and matplotlib have different order on color channel.

The different is:
- Matplotlib --> Red Green Blue
- Opencv --> Blue Green Red

So to solve this problem, we must convert the order channel. The code on note below:

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### Resize using ratio

Resize the image using the ratio method, which means the image has been resized based on a percentage ratio of the original image. For instance, as shown below

```python
w_ratio = 0.5
h_ratio = 0.5
img_resize = cv2.resize(img_rgb, (0,0), img_rgb, w_ratio,h_ratio)
```

```python
plt.imshow(img_resize)
```

```python
img_resize.shape
```

## Flip Image


Parameter of flip:
- Horizontal --> 0
- Vertical --> 1
- Both of them --> -1

```python
img_flip = cv2.flip(img_rgb, 0)
plt.imshow(img_flip)
```

```python
img_flip = cv2.flip(img_rgb, 1)
plt.imshow(img_flip)
```

```python
img_flip = cv2.flip(img_rgb, -1)
plt.imshow(img_flip)
```
